# 13. Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.


```python
df.columns
```

```python
df['dislike_rate']=df['dislikes']/df['view_count']*100
df['dislike_rate']
```

```python
import seaborn as sns
```

```python
pd.DataFrame(df[['dislike_rate','view_count']]).plot(x='view_count',y='dislike_rate',kind='scatter');
```

```python
s=df[['view_count','likes','dislikes','comment_count','dislike_rate']].corr()
```

```python
sns.heatmap(s.corr(),annot=True,cmap='viridis')
```

### INFERENCE:
    
    
   Scatter plot also follows a certain trend and pattern from left to right which shows there is a correlation using dislike rate
The correlation between views and dislikes is [0.67] which can be considered as moderate to higher correlation. So from this we can conclude that views have a signnificance in determining the dislikes of a video
